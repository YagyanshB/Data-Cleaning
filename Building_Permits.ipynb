```python
# Importing the required libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
# Uploading the dataset

df = pd.read_csv('Building_Permits.csv')
```

```python
# Inspecting a random sample of 10 rows within the Building Permits dataset

df.sample(10)
```

```python
# Getting the number of missing values within the dataset

missing_values_count = df.isnull().sum()
missing_values_count
```

```python
missing_values_count[7:43]
```

```python
# Total Values in our dataset

total_values = np.product(df.shape)
total_missing = missing_values_count.sum()

# Percentage of Data missing

(total_missing/total_values)*100
```

```python
# Getting a smaller subset of the data to fill in the missing values

subset_df = df.loc[:, 'Street Number Suffix': 'Location'].head()
subset_df
```

```python
# bfill is the backward fill method is where the values are filled from the bottom of the column
# the method starts from the bottom and moves upward
# if it encounters a missing value, it looks for the next valid (non-missing values)


subset_df.fillna(method = 'bfill', axis = 0).fillna(0)
```
